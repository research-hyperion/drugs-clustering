```python
df = pd.read_csv('drugs_age_interactions_clusters.csv')
#for item in DRUGBANK_DATABASES:
#    print(item[1])
```

```python
migrations = list()
for index, row in df.iterrows():
    drug_name = row['label']
    database_cluster = 0
    is_first_iteration = True
    drug_clusters_migration_dict = {"drug_name":drug_name, "migrations":list()}
    
    for item in DRUGBANK_DATABASES:
        cluster = row[(item[1]+" clusters")]
        if cluster != database_cluster:
            database_cluster = cluster
            if is_first_iteration:
                is_first_iteration = False
                drug_clusters_migration_dict["migrations"].append((item[1],database_cluster))
            else:
                drug_clusters_migration_dict["migrations"].append((item[1],database_cluster))
        else:
            drug_clusters_migration_dict["migrations"].append((item[1],database_cluster))
    migrations.append(drug_clusters_migration_dict)
                
        
```

```python
for item in migrations:
    print(item)
```

```python
csv_file = "migrations.csv"
with open(csv_file, mode='w') as clusters_file:
    clusters_write = csv.writer(clusters_file, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
    labels = ["Drug Name"]+[item[1] for item in DRUGBANK_DATABASES]
    clusters_write.writerow(labels)
    for item in migrations:
        csv_entry = list()
        csv_entry.append(item["drug_name"])
        for entry in item["migrations"]:
            csv_entry.append(entry[1])
        clusters_write.writerow(csv_entry)
```
